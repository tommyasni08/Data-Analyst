```python
### Import Python Package ###
import numpy as np
import pandas as pd
```

```python
### Import Raw Data ###
df = pd.read_csv("Data/us_congestion_2016_2022_sample_2m.csv")
```

```python
### Preprocess Data ###
## 1. Remove Invalid records ##
df = df[(~df['StartTime'].isnull()) & (~df['EndTime'].isnull()) & (df['StartTime']< df['EndTime']) ]
df = df[df['Distance(mi)'] > 0]

## 2. Convert datetime related data ##
df['StartTime'] = pd.to_datetime(df['StartTime'])
df['EndTime'] = pd.to_datetime(df['EndTime'])

## 3. Calculate the congestion duration ##
df['Duration'] = df['EndTime']-df['StartTime']
df['Duration'] = pd.to_timedelta(df['Duration']).dt.total_seconds()/60

## 4. Remove Invalid records part 2 ##
df = df[df['Duration']< 1000]

## 5. Reset Index ## 
df.reset_index(inplace=True, drop =True)
```
